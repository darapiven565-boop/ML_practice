**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]

arr.sort() # sorted 
n = len(arr)
if len(arr) % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = arr[n // 2 - 1] + arr[n // 2] /2

print(median_)
```

**Задача 2** \
Дані: [3, 3, 4, 100] \
Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100]

arr.sort() # sorted 
n = len(arr)
if len(arr) % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) /2

print(median_)

mean_ = sum(arr) / n
print(mean_)
```

**Задача 1** \
Дані: [10, 12, 12, 15, 20]\
Знайти моду.

```python
arr = [10, 12, 12, 15, 20]
# print(arr)
ans = {}
for num in arr:
    if num not in ans:
        ans[num] = 1
    else:
        ans[num] += 1
print(ans)
max_freq = 0
max_value = 0
for key, value in ans.items():
    if value > max_freq:
        max_freq = value
        max_value = key

print(max_value)
```

# Задача 1
Дані: [2, 4, 6, 8]

Потрібно:
- знайти середнє
- знайти квадрати відхилень
- знайти дисперсію
- інтерпретувати результат


```python
numbers = [2, 4, 6, 8]
len_numbers = len(numbers)
numbers.sort()
mean_numbers = sum(numbers) / len_numbers
```

```python
#квадрати відхилень
squared_deviations = []
for num in numbers:
    squared_deviations.append((num - mean_numbers) ** 2)
print(squared_deviations)
```

```python
variance_numbers = sum(squared_deviations) / len_numbers
```

```python
print(f'''
Mean is {mean_numbers}
Squared deviattions are {squared_deviations}
Varianse is {variance_numbers}
Mean and Variance are equal.
      ''')
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9] 

Середні однакові.  
Який набір стабільніший?

```python
a_set = [4, 5, 6]
b_set = [1, 5, 9]
```

```python
# рахуємо середнє значення
mean_a = sum(a_set) / len(a_set)
mean_b = sum(b_set) / len(b_set)
```

```python
#
a_set_step2 = []
for x in a_set:
    a_set_step2.append(x - mean_a)
print(a_set_step2)
```

```python
b_set_step2 = []
for x in b_set:
    b_set_step2.append(x - mean_b)
print(b_set_step2)
```

```python
a_step_3 = [i **2 for i in a_set_step2]
print(a_step_3)
b_step_3 = [j **2 for j in b_set_step2]
print(b_step_3)
```

```python
step_4_a = sum(a_step_3)
print(step_4_a)
step_4_b = sum(b_step_3)
print(step_4_b)
```

```python
ver_a = step_4_a / len(a_set)
ver_b = step_4_b / len(b_set)

print(f"a_set = {ver_a:.2f}, b_set = {ver_b:.2f}")
```
