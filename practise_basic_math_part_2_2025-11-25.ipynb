**Task 1.** У тебе є дані по користувачу за два дні:

- **День 1**: вектор `v1 = (перегляди, лайки, коментарі) = (10, 3, 1)`
- **День 2**: `v2 = (7, 5, 0)`

```python
import numpy as np

v1 = np.array([10, 3, 1])
v2 = np.array([7, 5, 0])

# use operation + in numpy 
# code
total = v1 + v2
print(total)
```

**Task 3.1**  З’ясувати, хто “сильніший” за загальною активністю.

Два користувачі: \
u1 = (50 переглядів, 10 лайків, 4 підписки) \
u2 = (70, 5, 1)



```python
import numpy as np

u1 = np.array([50, 10, 4])
u2 = np.array([70, 5, 1])

normalized_u1 = np.linalg.norm(u1)
normalized_u2 = np.linalg.norm(u2)
# use linalg.norm

print(normalized_u1, normalized_u2)
```
