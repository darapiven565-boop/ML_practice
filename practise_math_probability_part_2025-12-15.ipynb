```python
import pandas as pd
import numpy as np
```

```python
# друга команда краще, бо в залежності від задачі ж різна відповідь

data = [5, 10, 15, 60, 210] 
leangth = len(data)
variance_data = 0
mean_data = sum(data) / leangth
print(mean_data)
for num in data:
    result = (num - mean_data) ** 2
    # print(result)
    variance_data += result
    # pritnt(variance_data)
variance_data = variance_data / leangth
print(variance_data)
```

```python
std = variance_data ** 0.5
std
```

```python
np.random.seed(2)

n = 500
p = 0.10

A = stats.bernoulli.rvs(p, size=n)
B = stats.bernoulli.rvs(p, size=n)

# use stats.ttest_ind(B, A).pvalue,
# pvalue має бути менше 0.05 щоб результат вважався статистично значимим
p_value = stats.ttest_ind(B, A).pvalue
```

```python
import numpy as np
from scipy import stats


np.random.seed(42)

n = 500 

p_a = 0.10
p_b = 0.12

# use np.random.binomial to generate dist for p_a, p_b, with size n 

p_a_distr = stats.binom.rvs(p= p_a, n= 100, size= n)
sns.displot(p_a_distr)
p_b_distr = stats.binom.rvs(p= p_b, n= 100, size= n)
sns.displot(p_b_distr)
```

```python
# calculate mean, can use .mean()
mean_a = p_a_distr.mean()
print(mean_a)
mean_b = p_b_distr.mean()
print(mean_b)
```

```python
# create two variables and save result of stats.ttest_ind(<you group result B>, <you group result A>)
p_value_2 = stats.ttest_ind(p_b_distr, p_a_distr).pvalue
p_value_2
```
