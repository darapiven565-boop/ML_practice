```python
# Task 0.3. Dot transpose matrix . Calculate diagonal
import numpy as np

# | user | sessions | watch_hours | spend |
# | ---- | -------- | ----------- | ----- |
# | 1    | 5        | 10          | 100   |
# | 2    | 10       | 20          | 200   |
# | 3    | 3        | 6           | 60    |
# | 4    | 8        | 16          | 160   |

X = np.array([
    [5, 10, 100],
    [10, 20, 200],
    [3, 6, 60],
    [8, 16, 160]
])

# use .T and @
X1 = X.T @ X
print(1)
print(X1)

# use by prev task np.diag
print(2)
print(np.diag(X1))

# use np.linalg.matrix_rank 
print("--3--")
print(np.linalg.matrix_rank(X))
```

```python
# Calculate weights
weights = np.array([0.5, 0.2, 0.01])
impact = X * weights
impact
```

```python
# L2-normalize
np.linalg.norm(X[:, 0])  # sessions
np.linalg.norm(X[:, 1])  # watch_hours
np.linalg.norm(X[:, 2])  # spend
```
