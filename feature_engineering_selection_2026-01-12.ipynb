##### Обробка категоріальних ознак
ML-модель не розуміє текст. Вона розуміє лише числа.

country = "UA" \
contract = "Month-to-month" - не може напряму потрапити в модель

**Категоріальна ознака** — це колонка, яка:
> містить назви, а не числа \
> має обмежений набір значень


**One-Hot Encoding:**
Зробити окрему колонку для кожного можливого варіанту і відповідати так / ні

**Було:**
Contract \
Month-to-month \
One year \
Two year 

**Стало:** \
Contract_One year | Contract_Two year \
        0         |        0 \
        1         |        0 \
        0         |        1

Одна категорія -> одна колонка (ознака)

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import average_precision_score

from sklearn.linear_model import LogisticRegression
```

```python
y = df["Churn"].map({"Yes": 1, "No": 0})

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=["Churn"]),
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
num_cols = ["tenure", "MonthlyCharges", "TotalCharges"]
cat_cols = ["Contract", "InternetService", "PaymentMethod"]

X_train_1 = X_train[num_cols + cat_cols].copy()
X_test_1  = X_test[num_cols + cat_cols].copy()
```

```python
numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(drop="first", handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ],
    remainder="drop"
)

model = LogisticRegression(max_iter=500)

pipe1 = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", model)
])
```

```python
param_grid1 = [
    {
        "model__solver": ["liblinear"],
        "model__penalty": ["l1", "l2"],
        "model__C": [0.01, 0.1, 1, 10]
    },
    {
        "model__solver": ["lbfgs"],
        "model__penalty": ["l2"],
        "model__C": [0.01, 0.1, 1, 10]
    }
]

gs1 = GridSearchCV(
    pipe1,
    param_grid=param_grid1,
    scoring="accuracy",
    cv=5,
    n_jobs=-1,
    verbose=1
)

gs1.fit(X_train_1, y_train)
```
